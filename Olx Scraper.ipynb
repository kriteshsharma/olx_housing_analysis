```python
import scrapy
from scrapy.crawler import CrawlerProcess
import json
```

```python
class Olx(scrapy.Spider):
    name ='olx'
    
    url = "http://api.olx.in/relevance/v2/search?facet_limit=100&clientId=pwa&size=40&location_facet_limit=20&location=1000001&category=1725&lang=en-IN&clientVersion=8.25.0&user=17e3e06855cx84a4d9a&platform=web-desktop"
    
    headers = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36'
    }
    def __init__(self):
        with open('results.csv', 'w') as csv_file:
            csv_file.write('title,description,location,state,features,date,price\n')
            
    
    def start_requests(self):
        for page in range(0,50):
            
            yield scrapy.Request(url = self.url+ '&%page='+str(page) , headers= self.headers, callback=self.parse)
        
    def parse(self, res):
        data = res.text
        data = json.loads(data)
        
        for offer in data['data']:
            items = {
                'title' : offer['title'],
                'description' : offer['description'].replace('\n',' '),
                'location' :offer['locations_resolved']['ADMIN_LEVEL_1_name'],
                'state': offer['locations_resolved']['ADMIN_LEVEL_3_name'],
                'features': offer['main_info'],
                'date':offer['display_date'],
                'price':offer['price']['value']['display']
            }
            
            
            with open('results.csv', 'a',encoding="utf-8",newline='') as csv_file:
                writer = csv.DictWriter(csv_file, fieldnames=items.keys())
                writer.writerow(items)
#run spider
process = CrawlerProcess()
process.crawl(Olx)
process.start()

#debug
#lx.parse(Olx, '')
```
